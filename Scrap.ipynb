### Project 2  [Fetching Multiple List tags]

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
table = soup.find('div',{'class':'rankings-block__container full rankings-table'})
```

```python
count = 1
for data in table.find_all('span',{"class":"u-hide-phablet"}):
    print(count,"-->",data.text,"\n")
    count +=1
```

### Project 3  [Fetching Images]

```python
import requests 
from bs4 import BeautifulSoup
from PIL import Image
```

```python
url = 'https://www.harpersbazaar.com/culture/film-tv/g36039508/best-2021-movies/'
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
image = soup.find_all('img',{"alt":"best movies 2021"})
```

```python
filterImages = []
for img in image:
    try:
        filterImages.append(img['src'])
    except:
        filterImages.append(img['data-src'])
```

```python
filterImages
```

```python
cin = input("Enter Image Number (0,9) -->")
im = None
try:
    if int(cin)>10 or int(cin)<0:
        print("Enter Right Code")
    else:
        im = Image.open(requests.get(filterImages[int(cin)], stream=True).raw)
except:
    print("Enter Right Code")

im
```

### Project 4  [Scrapping Stack OverFlow]

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://stackoverflow.com/questions/415511/how-to-get-the-current-time-in-python"
```

```python
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
answers = soup.find_all('div',{"class":"answer"})
```

```python
for answer in answers:
    vote = answer.find('div',{"class":"js-vote-count"}).text
    if int(vote)>20:
        print(vote)
        codes = answer.find_all('pre')
        for code in codes:
            print(code.text)
    print("\n\n")
```
